# Missing Values

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({
    'missing_count': missing,
    'missing_pct': missing_pct
}).query('missing_count > 0').sort_values('missing_pct', ascending=False)

print(f"Columns with missing values: {len(missing_df)}")
print(f"\nTop 20 columns with most missing data:")
missing_df.head(20)
```

## Missing Value Analysis

The dataset contains 67 columns with missing values. 

Columns exceeding 40% missing data are dropped outright, as imputing the majority of a feature 
introduces more noise than signal and would likely harm model performance. 
These are predominantly property-related features (e.g. COMMONAREA, LIVINGAPARTMENTS, FLOORSMIN) 
which are sparsely recorded and less relevant to loan default risk.

For remaining columns under the 40% threshold, we will apply:
- **Median imputation** for numeric features — robust to outliers common in financial data
- **Mode imputation** for categorical features
- **Meaningful flags** where missingness carries information (e.g. OWN_CAR_AGE missing likely indicates no car ownership)

```python
# Drop columns with more than 40% missing
threshold = 40
cols_to_drop = missing_df[missing_df['missing_pct'] > threshold].index.tolist()
print(f"Columns to drop (>{threshold}% missing): {len(cols_to_drop)}")

df_clean = df.drop(columns=cols_to_drop)
print(f"Shape after dropping: {df_clean.shape}")
```

## Missing Value Imputation


`FLAG_OWN_CAR` already exists as a binary indicator for car ownership, so no additional 
feature engineering is needed for this field. We proceed directly to imputing remaining 
missing values across numeric and categorical columns.

- **Median imputation** for numeric columns — resistant to skewed distributions common in financial data
- **Mode imputation** for categorical columns — fills with the most frequent category

```python
# Separate numeric and categorical columns
numeric_cols = df_clean.select_dtypes(include=['float64', 'int64']).columns.tolist()
cat_cols = df_clean.select_dtypes(include=['object']).columns.tolist()

print(f"Numeric columns: {len(numeric_cols)}")
print(f"Categorical columns: {len(cat_cols)}")

# Median impute numeric
df_clean[numeric_cols] = df_clean[numeric_cols].fillna(df_clean[numeric_cols].median())

# Mode impute categorical
for col in cat_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])

# Verify no missing values remain
print(f"\nRemaining missing values: {df_clean.isnull().sum().sum()}")
```

# Target Variable Distribution

Before modelling, it is important to visually understand the class imbalance in our target 
variable. A heavily imbalanced dataset can cause a model to be biased towards the majority 
class (non-default), achieving high accuracy while completely failing to detect actual defaults — 
which is the most costly mistake in a lending context.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Count plot
sns.countplot(x='TARGET', data=df_clean, ax=axes[0], palette='Set2')
axes[0].set_title('Loan Default Distribution')
axes[0].set_xticklabels(['No Default (0)', 'Default (1)'])
axes[0].set_xlabel('')
for p in axes[0].patches:
    axes[0].annotate(f'{int(p.get_height()):,}', 
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='bottom', fontsize=11)

# Pie chart
axes[1].pie(df_clean['TARGET'].value_counts(), 
            labels=['No Default', 'Default'],
            autopct='%1.1f%%', 
            colors=sns.color_palette('Set2'),
            startangle=90)
axes[1].set_title('Default Rate')

plt.tight_layout()
plt.savefig('../data/processed/target_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print("Plot saved ✅")
```

## Anomaly Investigation — DAYS_EMPLOYED

The distribution of `DAYS_EMPLOYED` shows a suspicious spike at a large positive value. 
In this dataset, employment duration is recorded as negative days (days before application). 
A large positive value is nonsensical and likely represents a placeholder for missing or 
unknown employment status — commonly used to indicate pensioners or unemployed applicants. 
We investigate and engineer a flag to capture this.

```python
print(df_clean['DAYS_EMPLOYED'].describe())
print(f"\nSuspicious value (365243) count: {(df_clean['DAYS_EMPLOYED'] == 365243).sum()}")

# Flag anomalous value then replace with median
df_clean['DAYS_EMPLOYED_ANOMALY'] = (df_clean['DAYS_EMPLOYED'] == 365243).astype(int)
df_clean['DAYS_EMPLOYED'] = df_clean['DAYS_EMPLOYED'].replace(365243, np.nan)
df_clean['DAYS_EMPLOYED'] = df_clean['DAYS_EMPLOYED'].fillna(df_clean['DAYS_EMPLOYED'].median())

print(f"\nAnomaly flag distribution:\n{df_clean['DAYS_EMPLOYED_ANOMALY'].value_counts()}")
```

55,374 applicants had that anomalous value — that's 18% of the dataset. Definitely not random, almost certainly pensioners or unemployed.

# Feature Engineering

Raw features alone are often insufficient for capturing the financial relationships that 
drive loan default. We engineer domain-informed features that reflect real-world lending 
risk factors:

- **CREDIT_INCOME_RATIO** — how large the loan is relative to income; higher ratios indicate greater repayment burden
- **ANNUITY_INCOME_RATIO** — monthly repayment as a proportion of income; a key affordability metric used by lenders
- **CREDIT_TERM** — implied loan duration in months; longer terms carry more risk
- **DAYS_BIRTH_YEARS** — applicant age in years; more interpretable than raw negative days
- **DAYS_EMPLOYED_YEARS** — years in current employment; longer employment signals stability

These engineered features often rank among the most important predictors in credit risk models.

```python
# Credit to income ratio — key affordability metric
df_clean['CREDIT_INCOME_RATIO'] = df_clean['AMT_CREDIT'] / df_clean['AMT_INCOME_TOTAL']

# Annuity to income ratio — monthly repayment burden
df_clean['ANNUITY_INCOME_RATIO'] = df_clean['AMT_ANNUITY'] / df_clean['AMT_INCOME_TOTAL']

# Implied loan term in months
df_clean['CREDIT_TERM'] = df_clean['AMT_CREDIT'] / df_clean['AMT_ANNUITY']

# Age in years (DAYS_BIRTH is negative)
df_clean['AGE_YEARS'] = np.abs(df_clean['DAYS_BIRTH']) / 365

# Years employed
df_clean['YEARS_EMPLOYED'] = np.abs(df_clean['DAYS_EMPLOYED']) / 365

print("Engineered features added ✅")
print(df_clean[['CREDIT_INCOME_RATIO', 'ANNUITY_INCOME_RATIO', 
                 'CREDIT_TERM', 'AGE_YEARS', 'YEARS_EMPLOYED']].describe())
```

Summary:
- Average applicant is 44 years old, range 20–69
- Average loan is ~4x annual income — realistic for consumer lending
- Average annuity is 18% of income — monthly repayment burden
- Average loan term is ~22 months
- One outlier with credit ratio of 84x income — we may cap that later

# Correlation Analysis

We examine how strongly each numeric feature correlates with the target variable. 
This gives us a ranked view of which features have the most linear relationship with 
loan default, and helps validate whether our engineered features add predictive signal 
beyond the raw columns.

```python
# Correlation of numeric features with target
correlations = df_clean.select_dtypes(include=['float64', 'int64'])\
                        .corr()['TARGET']\
                        .drop('TARGET')\
                        .sort_values(key=abs, ascending=False)

# Plot top 20
plt.figure(figsize=(10, 8))
correlations.head(20).sort_values().plot(kind='barh', color=[
    'salmon' if x > 0 else 'steelblue' for x in correlations.head(20).sort_values()
])
plt.title('Top 20 Features Correlated with Loan Default', fontsize=13)
plt.xlabel('Pearson Correlation with TARGET')
plt.axvline(0, color='black', linewidth=0.8)
plt.tight_layout()
plt.savefig('../data/processed/feature_correlations.png', dpi=150, bbox_inches='tight')
plt.show()
print("Plot saved ✅")
```

key takeaways:

- AGE_YEARS & YEARS_EMPLOYED — our engineered features are already in the top 20, validating the feature engineering ✅
- DAYS_EMPLOYED_ANOMALY — our anomaly flag is also predictive, confirming the pensioner/unemployed hypothesis ✅
- REGION_RATING — higher risk regions correlate with default, interesting geographic signal
- Correlations are modest (~0.15 max) — expected for credit data, XGBoost will capture non-linear relationships much better

# Categorical Feature Analysis

We examine default rates across key categorical features to understand how loan type, 
gender, and income type relate to default risk. Rather than raw counts, we plot the 
default rate within each category — this is more meaningful given our class imbalance.

```python
cat_features = ['NAME_CONTRACT_TYPE', 'CODE_GENDER', 'NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, col in enumerate(cat_features):
    default_rate = df_clean.groupby(col)['TARGET'].mean().sort_values(ascending=False)
    default_rate.plot(kind='bar', ax=axes[i], color='salmon', edgecolor='black', linewidth=0.5)
    axes[i].set_title(f'Default Rate by {col}', fontsize=11)
    axes[i].set_ylabel('Default Rate')
    axes[i].set_xlabel('')
    axes[i].tick_params(axis='x', rotation=30)
    for p in axes[i].patches:
        axes[i].annotate(f'{p.get_height():.1%}',
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='bottom', fontsize=9)

plt.suptitle('Default Rate by Categorical Features', fontsize=13, y=1.01)
plt.tight_layout()
plt.savefig('../data/processed/categorical_default_rates.png', dpi=150, bbox_inches='tight')
plt.show()
print("Plot saved ✅")
```

Key insights:

- **Contract type** — Cash loans default at 8.3% vs 5.5% for revolving loans
- **Gender** — Males default at 10.1% vs 7.0% for females, a meaningful gap
- **Income type** — Maternity leave (40%) and Unemployed (36.4%) are extreme high-risk groups, while Pensioners and State servants are much safer
- **Education** — Clear inverse relationship, Lower secondary 10.9% down to Academic degree 1.8% — education level is a strong signal
